### EDA

```python
df.head()
```

que pais tiene las compañias mas grandes

```python
paises=[]
for i in range(len(df)):
    if df.iloc[i]['company_size']=='L':
        paises.append(((df.iloc[i]['company_location'])))

pd.DataFrame(paises).value_counts().head(5)
```

¿En que puesto se gana mas?

```python
df.groupby(['job_title']).agg({'salary_in_usd':'mean'}).round(2).sort_values(by='salary_in_usd',ascending=False)
```

Tipo de trabajo, pais de residencia donde mas se cobra?

```python
df.groupby(['employee_residence']).agg({'salary_in_usd':'mean'}).round(2).sort_values(by='salary_in_usd',ascending=False)
```

```python
df.groupby(['employment_type']).agg({'salary_in_usd':'mean'}).round(2).sort_values(by='salary_in_usd',ascending=False)
```

cuanto se cobra de media por cada nivel de experiencia?

```python
df.groupby(['experience_level']).agg({'salary_in_usd':'mean'}).round(2).sort_values(by='salary_in_usd',ascending=False)
```

en que pais se cobra  mas por tipo de empleo?

```python
df.groupby(['employment_type','company_location']).agg({'salary_in_usd':'mean'}).round(2).sort_values(by='salary_in_usd',ascending=False)
```
